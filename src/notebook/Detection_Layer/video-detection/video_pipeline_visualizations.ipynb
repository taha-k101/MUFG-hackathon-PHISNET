```python
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
from sklearn.metrics import confusion_matrix
from math import pi

# -------------------------------
# Prepare data
# -------------------------------
y_true = y_test
y_pred = clf.predict(X_test)
y_prob = clf.predict_proba(X_test)[:,1]

feature_cols = [col for col in df_video.columns if col != 'video']
feature_importances = clf.feature_importances_

epochs = list(range(1, len(clf.history['loss'])+1))
loss_values = clf.history['loss']
cumulative_energy = [epoch * X_train.shape[1] for epoch in epochs]

co2_per_epoch = 0.0005
co2_dynamic = np.array([co2_per_epoch*epoch for epoch in epochs])

# Multi-modality radar chart dummy values
metrics = ['Accuracy','Loss','Energy Proxy']
values_video = [0.68, 0.32, 13]
values_audio = [0.64, 0.36, 15]
values_fusion = [0.73, 0.27, 12]

values_video += values_video[:1]
values_audio += values_audio[:1]
values_fusion += values_fusion[:1]

angles = [n / float(len(metrics)) * 2 * pi for n in range(len(metrics))]
angles += angles[:1]

# -------------------------------
# Create figure with subplots
# -------------------------------
fig = plt.figure(figsize=(20,18))

# 1. Confusion Matrix
ax1 = plt.subplot(3,2,1)
cm = confusion_matrix(y_true, y_pred)
sns.heatmap(cm, annot=True, fmt='d', cmap='Blues', xticklabels=['Class 0','Class 1'], yticklabels=['Class 0','Class 1'], ax=ax1)
ax1.set_title("Confusion Matrix")
ax1.set_xlabel("Predicted")
ax1.set_ylabel("Actual")

# 2. Prediction Probability Distribution
ax2 = plt.subplot(3,2,2)
sns.histplot(y_prob[y_true==0], color='blue', label='Class 0', kde=True, stat="density", bins=15, ax=ax2)
sns.histplot(y_prob[y_true==1], color='red', label='Class 1', kde=True, stat="density", bins=15, ax=ax2)
ax2.set_title("Prediction Probability Distribution")
ax2.set_xlabel("Predicted Probability for Class 1")
ax2.set_ylabel("Density")
ax2.legend()

# 3. Feature Importance Barplot (Top 20)
ax3 = plt.subplot(3,2,3)
top_n = 20
top_indices = np.argsort(feature_importances)[-top_n:]
top_features = [feature_cols[i] for i in top_indices]
top_values = feature_importances[top_indices]
sns.barplot(x=top_values, y=top_features, palette="viridis", ax=ax3)
ax3.set_title("Top 20 Feature Importances (Agentic Reflection)")
ax3.set_xlabel("Importance")
ax3.set_ylabel("Features")

# 4. Feature Correlation Heatmap
ax4 = plt.subplot(3,2,4)
corr = df_video[feature_cols].corr()
sns.heatmap(corr, cmap="coolwarm", center=0, cbar_kws={'label': 'Correlation'}, ax=ax4)
ax4.set_title("Feature Correlation Map")

# 5. Loss & Energy Proxy Over Epochs
ax5 = plt.subplot(3,2,5)
ax5.plot(epochs, loss_values, marker='o', color='blue', label='Loss')
ax5.plot(epochs, np.array(cumulative_energy)/1000, marker='x', color='green', linestyle='--', label='Energy Proxy (k-units)')
ax5.set_title("Training Dynamics & Energy Proxy")
ax5.set_xlabel("Epoch")
ax5.set_ylabel("Loss / Energy Proxy")
ax5.legend()
ax5.grid(True)

# 6. Multi-Modality Radar Chart
ax6 = plt.subplot(3,2,6, polar=True)
ax6.plot(angles, values_video, color='blue', linewidth=2, label='Video')
ax6.fill(angles, values_video, color='blue', alpha=0.25)
ax6.plot(angles, values_audio, color='red', linewidth=2, label='Audio')
ax6.fill(angles, values_audio, color='red', alpha=0.25)
ax6.plot(angles, values_fusion, color='green', linewidth=2, label='Fusion')
ax6.fill(angles, values_fusion, color='green', alpha=0.25)
ax6.set_xticks(angles[:-1])
ax6.set_xticklabels(metrics)
ax6.set_title("Multi-Modality Comparison")
ax6.legend(loc='upper right', bbox_to_anchor=(0.1, 0.1))

plt.tight_layout()
plt.show()
```
